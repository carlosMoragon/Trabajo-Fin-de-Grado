```python
import pandas as pd
import matplotlib.pyplot as plt

def histogram(scores):
    counts, bins, patches = plt.hist(scores, bins=20, edgecolor='black')

    # Añadir los valores encima de cada barra
    for count, bin_left, bin_right in zip(counts, bins[:-1], bins[1:]):
        bin_center = (bin_left + bin_right) / 2
        plt.text(bin_center, count, str(int(count)), ha='center', va='bottom')

    # Personalizar el gráfico
    plt.title("Histograma de Score")
    plt.xlabel("Score")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

```

```python
#histogram(df_sn[df_sn["Score"] == 0.0]["Score"]) Todos son superiores a 0.0
histogram(df_sn[df_sn["Score"] != 0.0]["Score"])
print(df_sn[df_sn["Score"] >= 1.0]["Score"].count())
histogram(df_sn[df_sn["Score"] >= 1.0]["Score"])
```
